```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Sanitation.csv')
```

```python
plt.rcParams.update({'font.size': 10})
plt.figure(figsize=(80,40))
x = []
u = np.arange(len(df['Public Toilets']))
for name in df['ULB Name']:
    x.append(name)
ulb = len(x)
y = np.array(df['Public Toilets'])
y1 = np.array(df['She Toilets'])
y2 = np.array(df['Community Toilets'])
plt.xlabel("ULB(Urban Local Bodies)")
#plt.ylabel("Public Toilets")
plt.title("Public Toilets, she toilets & community toilets  per ULB")
plt.bar(u,y,color='red',width=0.5)
plt.bar(u+0.6,y1,color='g',width=0.5)
plt.bar(u+0.85,y2,color='b',width=0.5)
plt.legend(labels=['Public Toilets','She Toilets','Community Toilets'])
plt.xticks([i+0.5 for i in range(ulb)],x)
plt.show()
```
