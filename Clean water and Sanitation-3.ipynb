```python
Health_care_centers_frequency = {}
Urban_Health_centers = list(sanitation_df.get("Urban Health Centres"))
for i in Urban_Health_centers:
  try:
    Health_care_centers_frequency[i] += 1
  except:
    Health_care_centers_frequency[i] = 1
Health_care_centers_frequency
```

```python
x_axis = list(Health_care_centers_frequency.keys())
y_axis = list(Health_care_centers_frequency.values())
plt.xlabel("No.of hospital")
plt.ylabel("No.of districts")
plt.title("No.of cities vs no.of hospitals")
plt.bar(x_axis,y_axis)
plt.show()
```
